```python
# 1. 필요한 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt

# 2. 가상의 학생 성적 데이터 생성
data = {
    '학생': ['김민준', '이서연', '박도윤', '정수아', '최서우'],
    '수학': [90, 85, 95, 88, 92],
    '영어': [80, 92, 88, 95, 85]
}

df = pd.DataFrame(data)

# 3. 데이터프레임 내용 출력
print("학생 성적 데이터:")
print(df)

# 4. 막대 그래프로 시각화
# 한글 폰트 설정 (Mac) - 컴퓨터에 따라 폰트명이 다를 수 있습니다.
# 폰트 설정이 안 되면 한글이 깨져 보일 수 있습니다.
plt.rcParams['font.family'] = 'AppleGothic'  
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

# 그래프 생성
plt.figure(figsize=(8, 5))
plt.bar(df['학생'], df['수학'], label='수학 점수', color='skyblue')
plt.bar(df['학생'], df['영어'], label='영어 점수', color='salmon', alpha=0.7)

# 그래프 제목과 레이블 추가
plt.title('학생별 수학 및 영어 점수 비교')
plt.xlabel('학생 이름')
plt.ylabel('점수')
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.6)

# 그래프 보여주기
plt.show()
```
